```python
import pandas as pd 
import numpy as np
import plotly.express as px
import seaborn as sns
from sklearn.linear_model import LinearRegression
df = pd.read_csv('/home/gchen22/Desktop/maggiechen615/Unicorn_Companies.csv')
```

```python
df.head(5)
```

```python
import math
def split(v):
    return math.log(float(v.split('$')[1]))
```

```python
# Display and plot top 20 unicorn companies
dff = df.sort_values(by = 'Valuation ($B)', ascending=False)[:20]
dff['Valuation_Log'] = dff['Valuation ($B)'].apply(split)
dff['Founded Year'] = dff['Founded Year'].astype(int)
dff
```

```python
scat1 = px.scatter(
    dff, 
    x = 'Founded Year',
    y = 'Valuation_Log',
    color = 'Country',
    size = 'Valuation_Log',
    hover_name = 'Company')

scat1.show()
```

The graph indicates that within top 20 unicorns, firms that were founded before 2010 were all US companies, while 2010-2020 witnessed more unicorn companies rising in other parts of the world, e.g. India, China, Europe. 

Among the top 20 unicorn comps, Stripe has the highest valuation of ~$95B, which is around 10 times of the other companies (the others all around the range of $8-10B)

```python
# Deep dive into the emergence of unicorns over different time period

df = df.replace(to_replace='None', value=np.nan).dropna()
df['Founded Year'] = df['Founded Year'].astype(int)
df['Valuation_Log'] = df['Valuation ($B)'].apply(split)

fig = px.scatter(
    df, 
    x = 'Founded Year',
    y = 'Valuation_Log',
    color = 'Country',
    size = 'Valuation_Log',
    hover_name = 'Company')
    
fig.show()
    
```

While recent decade did witness more unicorns rising from countries other than U.S., U.S. still seems to be the country with the most unicorns, which can be further proved by the below plot:

```python
country_unicorn = df.groupby('Country').size()
px.bar(country_unicorn)
```

Lastly, besides the dimension of country, add on one dimension of "industry" with treemap. 
The treemap reveals that for the top 3 countries (US, UK, China), both US and UK are strong in Fintech. US is also strong in Internet Software and Services, while China is stronger in AI, Edtech, E-commerce and Hardware.

```python
fig = px.treemap(
    df, path = ['Industry', 'Country', 'Company'],
    values = 'Valuation_Log',
    color_discrete_sequence=px.colors.qualitative.Pastel)

fig.show()
```
